# 5.0 Building Machine Learning Models (Random Forest, LightGBM, XGBoost)

The reasons to use these 3 popular algorithms are:

**Random Forest algorithm:**
* Is less prone to overfitting than some other algorithms, such as decision trees, because it averages the predictions of multiple decision trees. 
* It can handle both numerical and categorical data, and it can handle missing values.

* It is a flexible algorithm that can be used for a wide range of tasks, including classification, regression, and feature selection.

**XGBoost:**

* **High predictive accuracy**: XGBoost has been shown to be highly accurate on a wide range of datasets, and it often outperforms other machine learning algorithms.

* **Scalability**: XGBoost is designed to be highly scalable and can handle large datasets with many features.

* **Feature importance**: XGBoost provides a way to measure the importance of each feature in the model, which can be useful for feature selection and understanding the underlying patterns in the data.

* **Regularization**: XGBoost uses regularization techniques such as L1 and L2 regularization to prevent overfitting and improve the generalization performance of the model.

* **Flexibility**: XGBoost provides a wide range of hyperparameters that can be tuned to optimize the model for a specific problem, making it a flexible algorithm that can be used for many different applications.

**LightGBM (Light Gradient Boosting Machine) is a gradient boosting framework that uses tree-based learning algorithms.**

* **Speed**: LightGBM is faster than other gradient boosting frameworks such as XGBoost and CatBoost, especially when dealing with large datasets.

* **Efficiency**: LightGBM uses a technique called "Histogram-based" which groups continuous values into discrete bins, which makes it more efficient in terms of memory and computation.

* **High accuracy**: LightGBM can handle a large number of features and can provide higher accuracy compared to other algorithms in some cases.

* **Customizable**: LightGBM provides many customizable parameters, which allows the user to fine-tune the model to fit specific data and requirements.

* **Parallel processing**: LightGBM supports parallel processing, which can significantly speed up the training process for large datasets.

## 5.2 Model evaluation function fits training data, predicts and computes score

```python
def model_eval(X_train, y_train, X_test, y_test, model):
    '''
    input: X is features data frame, y is response, model
    output: train cross validation score, test score, predicted values
    
    '''

    #X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=41)
    model.fit(X_train, np.ravel(y_train))
    y_pred = model.predict(X_test)
    score = cross_val_score(model, X_train, y_train, scoring = 'r2', cv = 5)    # Array of R^2 scores of the estimator for each run of the cross validation.
    test_score = r2_score(y_test, y_pred)             # R^2 (coefficient of determination) regression score function.            
    return score, test_score, y_pred

def plot_dist(col):
    '''
    plot distribution of feature in column col
    '''
    dfc = df[['id', col]].groupby([col]).count()
    (dfc/df.shape[0]).plot(kind="bar");

def plot_features(X, model, name):
    '''
    plot features importance of algoritm
    input: dataframe X and regression model
    output: Feature importance plot
    '''
    print("Plot feature importancesâ€¦")
    
    headers = ["name", "score"]
    values = sorted(zip(X.columns, model.feature_importances_), key=lambda x: x[1] * -1)
    forest_feature_importances = pd.DataFrame(values, columns = headers)
    forest_feature_importances = forest_feature_importances.sort_values(by = ['score'], ascending = False)
    features = forest_feature_importances['name'][:25]
    y_pos = np.arange(len(features))
    scores = forest_feature_importances['score'][:25]
    
    #plot feature importances
    plt.figure(figsize=(13,8))
    plt.barh(y_pos, scores, align='center', alpha=0.5, color='red')
    plt.yticks(y_pos, features, rotation=0)   
    plt.ylabel('Features')
    plt.xlabel('Score')
    plt.title('Feature importances ' + name)
    plt.show()
```

**Notes**:
The `model_eval` function you provided calculates two model evaluation metrics:

1. **Cross-Validation R-squared (RÂ²) Score (`score`)**:
   - This metric is calculated using the `cross_val_score` function, with the `scoring='r2'` parameter. It computes the RÂ² score (coefficient of determination) for your model during cross-validation. RÂ² measures the proportion of variance in the target variable that is explained by the model. It ranges from 0 to 1, with higher values indicating a better fit. The `score` variable will contain an array of RÂ² scores, one for each fold in the cross-validation.

2. **Test R-squared (RÂ²) Score (`test_score`)**:
   - This metric is calculated using the `r2_score` function, which computes the RÂ² score between the actual target values (`y_test`) and the predicted values (`y_pred`). It measures how well your model performs on the test set, with higher values indicating a better fit.

So, the `score` variable contains the cross-validated RÂ² scores, and the `test_score` variable contains the RÂ² score on the test data.

Keep in mind that RÂ² is just one of the possible evaluation metrics for regression models, and you might want to consider additional metrics depending on your specific problem and goals.


```python
# some columns will be excluded since these price related features are realistically not available when price needs to be forecasted
exclude=  ['id', 'price']

col = [i for i in col if i not in exclude]
print('Number of columns is:', len(col))
print(col)

# Next, set up the input features X_num and the target variable Y from the training dataset (df_train) 
# and the input features X_test and target variable Y_test from the test dataset (df_test).
# test = actual

X_num = df_train[col]
Y = df_train["price"]
Y_test = df_test["price"]
X_test = df_test[col]

# Creates a Random Forest regression model with 100 trees (n_estimators=100), 
# Use mean squared error as the criterion to split nodes (criterion='mse'), 
# Set the random state to 3 (random_state=3), 
# and uses all available processors for parallel computation (n_jobs=-1).

model = RandomForestRegressor(n_estimators=100,                
                               criterion='squared_error', 
                               random_state=3, 
                               n_jobs=-1)

# Call the model_eval() function, which takes the training and test data along with the model as input, 
# fits the model to the training data, 
# and computes the R2 score and the mean of cross-validation scores.

score, test_score, y_pred = model_eval(X_num, Y, X_test, Y_test, model)
# score: It is a list of cross-validation scores obtained from fitting the model to the training data. 
# Each score is the R2 score obtained for a particular fold of the cross-validation process.
# test_score: It is the R2 score obtained when using the trained model to predict the target variable on the test dataset.
# y_pred: It is an array of predicted target variable values obtained from using the trained model to predict the target variable on the test dataset.

print('\n')
print('Scenario RF with broad set of selected features:')
print('- Mean of crossvalidation scores:', np.mean(score))
print('- r2_score:', test_score)
```

**Noticings:**
The mean of cross-validation scores and R2 score are performance metrics used to evaluate the performance of a machine learning model.

The mean of cross-validation scores represents the average score of the model's performance across multiple subsets of the training data. It is a useful metric to avoid overfitting since it estimates how well the model can generalize to new data.

#### To avoid meaningless columns due to the One Hot Encoding wrapper function, which provides labels to the encoded (dummy) columns.

An elegant way to add labels to the One Hot Encoder output while keeping the â€œone lineâ€ implementation is to create a wrapper function around the OneHotEncoder class that can help to provide labels to the encoded columns.

This class inherits from the sklearn.preprocessing.OneHotEncoder, which means all the original functionality is preserved. Here you can find the [Towards Data Science article](https://towardsdatascience.com/how-to-assign-labels-with-sklearn-one-hot-encoder-e59a5f17df4f) where I copied the code from:

```python
from sklearn.preprocessing import OneHotEncoder as SklearnOneHotEncoder

class OneHotEncoder(SklearnOneHotEncoder):
    '''
    input: SklearnOneHotEncoder is the original sklearn.preprocessing.OneHotEncoder
    output: meaningless number labels are replaced by meaningfull column name labels
    '''
    def __init__(self, **kwargs):
        super(OneHotEncoder, self).__init__(**kwargs)
        self.fit_flag = False

    def fit(self, X, **kwargs):
        out = super().fit(X)
        self.fit_flag = True
        return out

    def transform(self, X, **kwargs):
        sparse_matrix = super(OneHotEncoder, self).transform(X)
        new_columns = self.get_new_columns(X=X)
        d_out = pd.DataFrame(sparse_matrix, columns=new_columns, index=X.index)
        return d_out

    def fit_transform(self, X, **kwargs):
        self.fit(X)
        return self.transform(X)

    def get_new_columns(self, X):
        new_columns = []
        for i, column in enumerate(X.columns):
            j = 0
            while j < len(self.categories_[i]):
                new_columns.append(f'{column}_<{self.categories_[i][j]}>')
                j += 1
        return new_columns
```

Tune XGBoost Regressor model to predict price via the hyperparameters:

Click [here](https://xgboost.readthedocs.io/en/latest/parameter.html) for source.

**n_estimators:** number of trees

**learning_rate:** Step size shrinkage used in update to prevents overfitting. After each boosting step, we can directly get the weights of new features, and eta shrinks the feature weights to make the boosting process more conservative.

**gamma:** Minimum loss reduction required to make a further partition on a leaf node of the tree. The larger gamma is, the more conservative the algorithm will be.

**subsample** Subsample ratio of the training instances. Setting it to 0.5 means that XGBoost would randomly sample half of the training data prior to growing trees. and this will prevent overfitting. Subsampling will occur once in every boosting iteration. Default is 1.

**colsample_bytree:** colsample_bytree is the subsample ratio of columns when constructing each tree. Subsampling occurs once for every tree constructed.

**max_depth:** Maximum depth of a tree. Increasing this value will make the model more complex and more likely to overfit. 0 is only accepted in lossguided growing policy when tree_method is set as hist and it indicates no limit on depth. Beware that XGBoost
