# 2.6.1 Drop Columns with Many Missing Values

```python
# Set most_missing_cols to calculate columns in df where more than 70% of values are missing or NaN
# df.isnull().sum() returns the number of missing values in each column
# abs() returns absolute values of missing values counts to ensure only high percentage of missing values are stored in missing_most_cols
most_missing_cols = set(df.columns[abs(df.isnull().sum()) > 0.7*len(df)])
print(most_missing_cols)

# Use drop() to drop most_missing_cols
df = df.drop(most_missing_cols, axis=1)
```

```python
# Confirm most_missing_cols have been dropped
df.columns
```

```python
# Create an extra feature with the number of bathrooms from bathroom_text
# Select first row of bathrooms_text column to see example of data in column
df['bathrooms_text'][0]
```

```python
# Use regular expression '(\d+), where '\d' matches any digit 
#'+' indicates that it should match one or more digits
# The parentheses '()' indicated that the matched digits should be captured and extracted by the 'str.extract()' method
# The extracted intergers are then stored in a new column named bathrooms in the dataframe.
df['bathrooms'] = df['bathrooms_text'].str.extract('(\d+)')
df['bathrooms'].head()
```

```python
# Make df['bathrooms'] a float dtype
df['bathrooms'] = df['bathrooms'].astype(float)
df['bathrooms'].dtypes
```

```python
# Check consise summary information of df
df.info()
```

# 2.7 Process Numerical Features

```python
# Combine 'bedrooms' and 'bathrooms' to reduce the number of features
df['total_rooms'] = df['bedrooms'] + df['bathrooms']
df['total_rooms'].head()
```

```python
# Use .apply(lambda X: float(str(x).replace('%', "")))
# First convert data to string using str()
# apply lambda function to every data in the target column
# using replace() to remove the % symbol and replace with empty string
# Use float to convert the str dtype into a float dtype
df['host_response_rate'] = df['host_response_rate'].apply(lambda x: float(str(x).replace('%', "")))
df['host_acceptance_rate'] = df['host_acceptance_rate'].apply(lambda x: float(str(x).replace('%', "")))

df[['host_response_rate', 'host_acceptance_rate']].head()
```

# 2.7.1 Drop Unnecessary Columns

```python
# Exclude 'host_listings_count' because it is the same as "host_total_listings_count" and it gives the same result
# Exclude other unnecessary columns such as 'scrape_id', 'amenities', 'bathrooms_text', 'host_id'
df = df.drop(columns=['host_listings_count', 'scrape_id', 'amenities', 'bathrooms_text', 'host_id'])
df.info()
```

```python
# Select all columns in df with float, int64 or int dtypes and assign to col_features
col_num_featues = df.select_dtypes(include = ['float', 'int', 'int64']).columns

# Use list comprehension to iterate through each element 'i' not in ['id', 'scrape_id']
col_num_features = [i for i in col_num_featues if i not in ['id', 'scrape_id']] 
print("The following are numeical features:\n", col_num_features)
```

# 2.7.2  Replacing Missing Values with The Mean

```python
# Define lambda function fill_mean
# Use fillna() pandas meanthod to fill any NaN with the mean
# mean() calculates the mean of the non-missing values in the column
# col is taking the place of x i.e apply(lambda x: :....)
fill_mean = lambda col: col.fillna(int(col.mean()))

# Fill the mean in the rows where there are NaNs
df[col_num_features] = df[col_num_features].apply(fill_mean, axis=0)

# Check if NaNs have been filled in with the mean using df[col_reviews].isnull().sum()
df[col_reviews].isnull().sum()
```

There are no missing values in the review columns or in col_num_features
