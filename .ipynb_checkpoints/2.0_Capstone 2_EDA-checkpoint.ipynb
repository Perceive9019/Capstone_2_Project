# 3.1 Which months and neighbourhoods have the highest prices?

```python
# Remove duplicates from dfcal
dfcal.drop_duplicates()

# Create new df called avg_monthly_price. Use .groupby() to group 'price' column by 'month' and calculate the mean price for each month
avg_monthly_price = dfcal[['month', 'price']].groupby('month').mean()

# Plot
plt.figure(figsize = (10,8))
plt.plot(avg_monthly_price['price'])
plt.ylabel('Mean Daily Prices ($)')
plt.xlabel('Month')
# Set the x-axis tick locations to every integer between 0 and 11 (inclusive) using plt.xticks(np.arange(0, 12, step=1)).
plt.xticks(np.arange(0, 12, step=1))
# Set the x-axis tick labels to the names of the months
plt.xticks(np.arange(12), ('Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'), rotation=45)
plt.title('Mean Daily Prices over 12 Months')

# Save plot to file called average price for month using plt.savefig()
plt.savefig('average price for month')

plt.show()
```

**Answer:** In San Diego the highest rental prices are during the month of August. We can confirm this by also looking at the number of listings during August.

```python
# Use groupby to return a DataFrame called "price_per_month_neighbourhood", which shows the mean value of each column of "df_merged" for each unique combination of "month" and "neighbourhood_cleansed". 
#This is useful for analyzing the relationship between the price of Airbnb listings and the month and neighborhood in which they are located.
price_per_month_neighbourhood = df_merged.groupby(['month', 'neighbourhood_cleansed']).mean().reset_index()
price_per_month_neighbourhood.head(2)
```

The diagram above shows the neigborhoods with the highest average prices. Unfortunately due to the number of neighbourhoods, it is difficult to make out which neighborhood had the highest average price. Therefore, we need to visually represent this data in another way that is much clearer.

```python
# #import seaborn as sns

# # Set figure size
# sns.set(rc={'figure.figsize':(15,8)})

# # Create line plot using Seaborn
# sns.lineplot(data=price_per_month_neighbourhood, x='month', y='price', hue='neighbourhood_cleansed')

# # Set x and y axis labels and title
# plt.xlabel('Month')
# plt.ylabel('Mean price in neighbourhood ($)')
# plt.title('Mean daily price depending on month of year and neighbourhood')

# # Move the legend below the plot
# plt.legend(bbox_to_anchor=(0.5,-0.2), loc="upper center", ncol=6)


# import seaborn as sns
# import matplotlib.pyplot as plt

# Set figure size
sns.set(rc={'figure.figsize':(18, 36)})  # Adjust the values as needed

# Get unique neighborhoods
unique_neighbourhoods = price_per_month_neighbourhood['neighbourhood_cleansed'].unique()

# Split neighborhoods into groups of 17
neighbourhood_groups = [unique_neighbourhoods[i:i+17] for i in range(0, len(unique_neighbourhoods), 17)]

# Create subplots
num_subplots = len(neighbourhood_groups)
fig, axes = plt.subplots(num_subplots, 1, sharex=True, figsize=(18, 6 * num_subplots))

# Create line plots for each subplot
for idx, neighbourhoods in enumerate(neighbourhood_groups):
    ax = axes[idx]
    for neighbourhood in neighbourhoods:
        data_neighbourhood = price_per_month_neighbourhood[price_per_month_neighbourhood['neighbourhood_cleansed'] == neighbourhood]
        sns.lineplot(data=data_neighbourhood, x='month', y='price', ax=ax, label=neighbourhood)
    
    ax.set_ylabel('Mean price in neighbourhood ($)')
    ax.legend(title='Neighbourhood')
    ax.set_title(f'Mean daily prices - Subplot {idx + 1}')

# Set x axis label and title for the last subplot
axes[-1].set_xlabel('Month')
axes[-1].set_title('Mean daily prices in neighbourhoods')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

The neighbourhoods most-likely with the highest rental prices are: **Gaslamp Quarter** (\~\$850), **La** **Jolla** (\~\$850), **Misson Bay** (\~\$690), **Old Town** (\~\$650), **LittleItaly** (\\$600 - \\$780), and **Pacific Beach** (\\$500 - \\$600). This conclusion is supported by the fact that these neigbourhoods also has high number of listings. Prices rose sharpley in July or between month 6 and month 8.
